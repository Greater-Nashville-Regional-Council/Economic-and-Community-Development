```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 150)
```

```python
data = pd.read_csv('powercom_people_spring23.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
data = data.loc[data['industry'] != 'not_working']
```

```python
data['naics'] = data['industry'].str.slice(start = 5, stop = 7)
```

```python
data.head()
```

```python
df = data[data['work_cty'] != 'Trousdale County, TN']
```

```python
tr = data[data['work_cty'] == 'Trousdale County, TN']
```

```python
outcommuters = pd.DataFrame(df['naics'].value_counts()).reset_index(drop = False)
outcommuters.head()
```

```python
outcommuters.to_csv('outcommuters.csv', index = False)
```

```python
stayers = pd.DataFrame(tr['naics'].value_counts()).reset_index(drop = False)
stayers.head()
```

```python
stayers.to_csv('stayers.csv', index = False)
```
