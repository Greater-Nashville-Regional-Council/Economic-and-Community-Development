This is data requested to give an overview of the EDA's Equity investment priority for the counties in the GNRC working region. The data is simply collected here and written out to create a simple Excel workbook.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from functools import reduce
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 150)
from functools import reduce
pd.options.mode.chained_assignment = None  # default='warn'
from warnings import simplefilter
simplefilter(action="ignore", category=pd.errors.PerformanceWarning)
import sys
sys.path.append("../../Functions and Dictionaries/") # Adds higher directory to python modules path
import geodict
namestocommon = geodict.namestocommon
geotogeoid = geodict.geotogeoid
tofullcensus = geodict.tofullcensus
import sqlite3 as sq
```

```python
#functions
def percent(x, y):
        try:
            return ((x/y)*100)
        except ZeroDivisionError:
            return 0
def percentchange(x, y):
    try:
        return ((x - y)*100/y)
    except ZeroDivisionError:
        return 0
def realchange(x, y):
    return x-y
```

```python
conn = sq.connect('../../Data-Pipelines/Outputs/CensusBureau.db')
#PL2020
sql_query = pd.read_sql('SELECT * FROM [CensusBureau_PL_2020]', conn)
PL2020 = pd.DataFrame(sql_query)
PL2020.drop(columns = 'Source', inplace = True)
#ACS52021
sql_query = pd.read_sql('SELECT * FROM [CensusBureau_ACS5_2021_1]', conn)
ACS52021_1 = pd.DataFrame(sql_query)
sql_query = pd.read_sql('SELECT * FROM [CensusBureau_ACS5_2021_2]', conn)
ACS52021_2 = pd.DataFrame(sql_query)
ACS52021 = ACS52021_1.merge(ACS52021_2, on = ['NAME', 'GEO_ID', 'Source'])
ACS52021.drop(columns = 'Source', inplace = True)
dfs2021 = [ACS52021, PL2020]
twothousandtwentyone = reduce(lambda  left,right: pd.merge(left,right,on=['NAME', 'GEO_ID'], how='outer'), dfs2021)
```

```python
data = twothousandtwentyone
```

```python
geos = ['Cheatham County', 'Davidson County', 'Dickson County', 'Houston County', 'Humphreys County', 
        'Stewart County', 
        'Sumner County', 'Williamson County', 'Wilson County', 'Tennessee', 'US']
```

```python
data = data.reset_index().set_index('NAME').transpose()
data = data.rename(columns = namestocommon)
data = data[geos]
data = data.transpose().reset_index()
data = data.drop(columns = 'index')
```

```python
cols = ['NAME', 'Population', 'White Alone', 'White Alone %', 'Non-White Population: Race',
        'Black or African American Alone', 'Black or African American Alone %', 'Black or African American Alone % Non-White Population', 
        'American Indian Alaska Native Alone', 'American Indian Alaska Native Alone %', 'American Indian Alaska Native Alone % Non-White Population', 
        'Asian Alone', 'Asian Alone %', 'Asian Alone % Non-White Population', 
        'Native Hawaiian Other Pacific Islander Alone', 'Native Hawaiian Other Pacific Islander Alone %', 
        'Native Hawaiian Other Pacific Islander Alone % Non-White Population', 
        'Some Other Race Alone', 'Some Other Race Alone %', 'Some Other Race Alone % Non-White Population', 
        'Two or More Races', 'Two or More Races %', 'Two or More Races % Non-White Population', 
        'Hispanic or Latino', 'Hispanic or Latino %', 'Not Hispanic or Latino', 'Not Hispanic or Latino %', 
        'Minority', 'Minority %', 'White Alone Not Hispanic or Latino', 'White Alone Not Hispanic or Latino %', 
        'Male', 'Male %', 'Female', 'Female %', 'Disability:With Disability', 'Disability%:With Disability', 
        'Disability:Hearing Difficulty', 'Disability%:Hearing Difficulty', 
        'Disability:Vision Difficulty', 'Disability%:Vision Difficulty', 
        'Disability:Cognitive Difficulty', 'Disability%:Cognitive Difficulty', 
        'Disability:Ambulatory Difficulty', 'Disability%:Ambulatory Difficulty', 
        'Disability:Self-Care Difficulty', 'Disability%:Self-Care Difficulty', 
        'Disability:Independent Living Difficulty', 'Disability%:Independent Living Difficulty', 
        'Disability:Total Disabilities', 'Disability:Population with One Disability', 
        'Disability%:Population with One Disability', 'Disability:Population with Two or More Disabilities', 
        'Disability%:Population with Two or More Disabilities'
       ]
data = data[cols]
```

```python
data['Asian, Native Hawaiian, Other Pacific Islander Alone'] = data['Asian Alone'] + data['Native Hawaiian Other Pacific Islander Alone']
data['Asian, Native Hawaiian, Other Pacific Islander Alone %'] = percent(data['Asian, Native Hawaiian, Other Pacific Islander Alone'], data['Population'])
data['Asian, Native Hawaiian, Other Pacific Islander Alone % Non-White Population'] = percent(data['Asian, Native Hawaiian, Other Pacific Islander Alone'], data['Non-White Population: Race'])
data['Other Persons of Color'] = data['Some Other Race Alone'] + data['Two or More Races']
data['Other Persons of Color %'] = percent(data['Other Persons of Color'], data['Population'])
data['Other Persons of Color % Non-White Population'] = percent(data['Other Persons of Color'], data['Non-White Population: Race'])
data['Other Persons of Color %'] = percent(data['Other Persons of Color'], data['Population'])
```

```python
data.to_csv('data/InvestmentPriorityEquity.csv', index = False)
```
