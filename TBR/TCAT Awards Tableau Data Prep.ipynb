```python
import pandas as pd
import sys
sys.path.append("../../Functions and Dictionaries/") # Adds higher directory to python modules path
import CIPCodes
programfam = CIPCodes.programfam
programcat = CIPCodes.programcat
program = CIPCodes.program
```

```python
data = pd.read_csv('data/TCAT_AwardsbyLocation.csv', dtype = str)
```

```python
data.head(40)
```

```python
data.info()
```

```python
inv_programfam = {v: k for k, v in programfam.items()}
inv_programcat = {v: k for k, v in programcat.items()}
inv_program = {v: k for k, v in program.items()}
```

```python
data['Program Family'] = data['2 Digit CIP'].map(inv_programfam)
data['Program Category'] = data['4 Digit CIP'].map(inv_programcat)
data['Program'] = data['6 Digit CIP'].map(inv_program)
```

```python
data = data.drop(columns = ['4 Dig', '6 Dig'])
```
